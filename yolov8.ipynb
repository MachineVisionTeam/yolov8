```python
import yaml

try:
    from yaml import CLoader as Loader, CDumper as Dumper
except ImportError:
    from yaml import Loader, Dumper
```

```python
with open("/content/drive/MyDrive/yolov8/data.yaml") as file:
  config = yaml.load(file.read(), Loader=Loader)
new_config = {
    "nc": config["nc"],
    "names": {
        i: v for i, v in enumerate(config["names"])
    },
    "train": "train/images",
    "val": "val/images",
    "path": "/content/drive/MyDrive/yolov8/"
}

with open("/content/drive/MyDrive/yolov8/datav8.yaml", "w") as file:
  file.write(yaml.dump(new_config, Dumper=Dumper))
```

```python
my_new_model = YOLO('/content/drive/MyDrive/yolov8/results/100_epochs-/weights/last.pt')
metrics = my_new_model.val(data="/content/drive/MyDrive/yolov8/datav8.yaml")
```
